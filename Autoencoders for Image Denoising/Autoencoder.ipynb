# Implementing Autoencoders to perform Image Denoising

## Importing Libraries


```python
import numpy as np

from tensorflow.keras.datasets import mnist
from matplotlib import pyplot as plt
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.callbacks import EarlyStopping, LambdaCallback
from tensorflow.keras.utils import to_categorical

%matplotlib inline
```

## Data Preprocessing


```python
#loading data
(x_train, y_train), (x_test, y_test) = mnist.load_data()  
#normalize values to 0 to 1 from 0 to 255
x_train=x_train.astype('float')/255.
x_test=x_test.astype('float')/255.
#reshape examples
x_train = np.reshape(x_train, (60000, 784))  #28*28=78 pixels
x_test = np.reshape(x_test, (10000, 784))  #28*28=78 pixels

```

## Adding Noise to dataset

```python
#adding noise to dataset
x_train_noisy= x_train + np.random.rand(60000,784)*0.9  #each value in x_train +some random value and scale it down by 0.9
x_test_noisy= x_test + np.random.rand(10000,784)*0.9  #each value in x_train +some random value and scale it down by 0.9

#to ensure values remain between 0 and 1, we do this
x_train_noisy= np.clip(x_train_noisy, 0., 1.)
x_test_noisy= np.clip(x_test_noisy, 0., 1.)
```

```python
#create plot function to plot first 10 images of a dataset
def plot(x, p, labels=False):  #p is prediction, x is dataset
    plt.figure(figsize=(20,2))  #we create a fig
    for i in range(10):
        #we create a subplot of 10 images with index i+1
        plt.subplot(1, 10, i+1)
        #we reshape since we flattened it befroe
        plt.imshow(x[i].reshape(28,28), cmap='binary')  #setting smap to binary shows black and white images
        #we remove ticks by setting it to blank lists
        plt.xticks([])
        plt.yticks([])
        if labels:
            plt.xlabel(np.argmax(p[i]))  #predictions will be displayed as labels of images
    plt.show()  #to display entire plot
        
        
    
    
plot(x_train, None)
```

```python
#same examples with noise added to it
plot(x_train_noisy, None)
```

##  Building and Training a Classifier

```python
#building a classifier: simple 2layer NN
classifier= Sequential([
    Dense(256, activation='relu', input_shape=(784,)),
    Dense(256, activation='relu'),
    Dense(10, activation='softmax')
    
])

#compile
classifier.compile(
optimizer='adam',
loss='sparse_categorical_crossentropy',  #we use this coz our labels are not encoded (one-hot), they are just numerical vals
metrics=['accuracy']    

)

#train classifier
classifier.fit(x_train, y_train, batch_size=512, epochs=3) 
```

```python
loss, acc=classifier.evaluate(x_test, y_test) #classifier works well on no noise added data 
print(acc)
```

```python
loss, acc= classifier.evaluate(x_test_noisy, y_test)  #classifier does not work on noisy data
print(acc)
```

## Implementing the Autoencoder

```python
input_image= Input(shape=(784,))
#encoding input
encoded = Dense(64, activation='relu')(input_image)  #forcing input_image to be reduced to 64 dim
#decoding output
decoded=Dense(784, activation='sigmoid')(encoded)  #expanding it back to 784 dim

autoencoder= Model(input_image, decoded)
autoencoder.compile(loss='binary_crossentropy', optimizer='adam')
```

## Training the Autoencoder

```python
autoencoder.fit(
x_train_noisy, x_train, epochs=100,
batch_size=512, validation_split=0.2, verbose=False,
callbacks=[
    EarlyStopping(monitor='val_loss', patience=5),
    LambdaCallback(on_epoch_end=lambda e, l: print('{:.3f}'.format(l['val_loss']), end=' _ '))
])

print(' _ ')
print('Training is complete!')
```

## Prediction using Autoencoders

```python
#make predictions on noisy test data in which you were earlier getting very bad accuracy
predictions = autoencoder.predict(x_test_noisy)
```

```python
#plot the noisy test images to make comparison
plot(x_test_noisy, None)
```

```python
#plot predictions for noisy test images
plot(predictions, None)
```

```python
#now see after denoising, how does your classifier now fare?
loss, acc = classifier.evaluate(predictions, y_test)
print(acc)
```

## Creation of a Composite Model which incorporates Autoencoder and your classifier!

```python
#create compositing model: input: noisy or non-noisy image, model will first denoise it then use it on the classifier to predict
input_image= Input(shape=(784,))
x= autoencoder(input_image)  #get denoised image from autoecnoder
y =classifier(x) # get prediction from classifier

#create model by putting it together
denoise_and_classify = Model(input_image, y)
```

```python
#now get predictions using your combinel model
predictions=denoise_and_classify.predict(x_test_noisy)
```

```python
#plot noisy_test_data with their predictions and make lables=True 
plot(x_test_noisy, predictions, True)
```

```python
#compare with orig labels
plot(x_test, to_categorical(y_test), True)
```
